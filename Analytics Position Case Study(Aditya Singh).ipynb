 ## 	Part A - Calculating loyalty points

```python
#function for filtering dataframe according to date and slot
def filtering_data(df,date_,S):
    date_ = pd.to_datetime(date_).date()
    res = df[(df['Datetime'].dt.date == date_) & (df['Slot'] == S)]
    return res
```

```python
#function for calculating Loyalty points 
def Loyalty_points(A,B,C):
    Games_Played = A.groupby('User ID')['Games Played'].sum()                 #sum of games played by each player
    Deposit_Amount = B.groupby('User Id')['Amount'].sum()                     #sum of deposited amount by each player
    Withdrawal_Amount = C.groupby('User Id')['Amount'].sum()                  #sum of withdrawal amount by each player
    number_deposits = B.groupby('User Id')['Amount'].count()                  #number of deposits per player
    number_withdrawals = C.groupby('User Id')['Amount'].count()               #number of withdrawals per player
    df = pd.concat([Games_Played,Deposit_Amount,Withdrawal_Amount,number_deposits,number_withdrawals], axis = 1)
    df.columns = ['Games Played', 'Deposit Amount', 'Withdrawal Amount', 
                  'Number of Deposits', 'Number of Withdrawals']
    #Converting data types to numeric as i was getting some errors dur to it.
    
    df['Deposit Amount'] = pd.to_numeric(df['Deposit Amount'], errors='coerce')
    df['Withdrawal Amount'] = pd.to_numeric(df['Withdrawal Amount'], errors='coerce')
    df['Games Played'] = pd.to_numeric(df['Games Played'], errors='coerce')
    df['Number of Deposits'] = pd.to_numeric(df['Number of Deposits'], errors='coerce')
    df['Number of Withdrawals'] = pd.to_numeric(df['Number of Withdrawals'], errors='coerce')
    
    df.fillna(0,inplace=True)
    Loyalty_Points = (0.01 * df['Deposit Amount']) + (0.005 * df['Withdrawal Amount']) + (0.001 * (df['Number of Deposits'] - df['Number of Withdrawals']).clip(lower=0)) + (0.2 * df['Games Played'])
    df['Loyalty Points'] = Loyalty_Points
    return df
```

2. Calculate overall loyalty points earned and rank players on the basis of loyalty points in the month of October. 
     In case of tie, number of games played should be taken as the next criteria for ranking.

```python
def filtering_data2(df,month_):
    res = df[(df['Datetime'].dt.month == month_)]
    return res
```

```python
def Loyalty_points2(A,B,C):
    Games_Played = A.groupby('User ID')['Games Played'].sum()                 #sum of games played by each player
    Deposit_Amount = B.groupby('User Id')['Amount'].sum()                     #sum of deposited amount by each player
    Withdrawal_Amount = C.groupby('User Id')['Amount'].sum()                  #sum of withdrawal amount by each player
    number_deposits = B.groupby('User Id')['Amount'].count()                  #number of deposits per player
    number_withdrawals = C.groupby('User Id')['Amount'].count()               #number of withdrawals per player
    df = pd.concat([Games_Played,Deposit_Amount,Withdrawal_Amount,number_deposits,number_withdrawals], axis = 1)
    df.columns = ['Games Played', 'Deposit Amount', 'Withdrawal Amount', 
                  'Number of Deposits', 'Number of Withdrawals']
    
    df['Deposit Amount'] = pd.to_numeric(df['Deposit Amount'], errors='coerce')
    df['Withdrawal Amount'] = pd.to_numeric(df['Withdrawal Amount'], errors='coerce')
    df['Games Played'] = pd.to_numeric(df['Games Played'], errors='coerce')
    df['Number of Deposits'] = pd.to_numeric(df['Number of Deposits'], errors='coerce')
    df['Number of Withdrawals'] = pd.to_numeric(df['Number of Withdrawals'], errors='coerce')
    
    df.fillna(0,inplace=True)
    Loyalty_Points = (0.01 * df['Deposit Amount']) + (0.005 * df['Withdrawal Amount']) + (0.001 * (df['Number of Deposits'] - df['Number of Withdrawals']).clip(lower=0)) + (0.2 * df['Games Played'])
    df['Loyalty Points'] = Loyalty_Points
    #till here its same as previous one
    df = df.sort_values(by=['Loyalty Points','Games Played'] , ascending = [False,False])
    df['Rank'] = df['Loyalty Points'].rank(method='dense', ascending=False)
    #df['Rank'] = range(1, len(df) + 1)
    return df
```
